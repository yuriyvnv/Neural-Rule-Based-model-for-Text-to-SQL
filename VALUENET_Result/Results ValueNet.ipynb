#### Process the ground truth queries and extract the ones that correspond to our dev set ("poker_player", "singer", "orchestra", "voter_1")

```python
import csv

# Open the text file and read the contents into a list of lines
with open("/Users/yuriyperezhohin2/Desktop/MastersAwareTech/Masters/ThesisImplementation/GitLab/sandabox-yuriy/VALUENET/ground_truthTrainVALUENET.txt", "r") as f:
    lines = f.readlines()

# Create a list of rows, where each row is a list of the three columns
rows = []
for line in lines:
    # Split the line on the tab character
    parts = line.split("\t")

    # Create a row from the three parts
    row = [parts[0], parts[1], parts[2]]

    # Add the row to the list of rows
    rows.append(row)

# Create a CSV writer object
with open("ground_truthValueNetTrain.csv", "w", newline="") as f:
    writer = csv.writer(f)

    # Write the header row
    writer.writerow(["SQL", "database", "question"])

    # Write the data rows
    for row in rows:
        writer.writerow(row)
```

```python
import pandas as pd
dev_set = pd.read_csv("/Users/yuriyperezhohin2/Desktop/MastersAwareTech/Masters/ThesisImplementation/GitLab/sandabox-yuriy/VALUENET/ground_truthValueNet.csv")
```

```python
databases = ["book_2", "gymnast", "musical", "farm","device", "race_track", "ship_1", "school_bus", "wrestler", "department_management", "climbing","journal_committee", "body_builder","performance_attendance", "election_representative"]
specificSQLDatabases=[]
specificQuestionDatabases= []
indexes= []
for index , row in dev_set.iterrows():
    if (row["database"] in databases )& ("union" not in str.lower(row["SQL"])) & ("intersect" not in str.lower(row["SQL"])) & ("except" not in str.lower(row["SQL"])) :
        specificQuestionDatabases.append(row["question"])
        specificSQLDatabases.append(row["SQL"])
        indexes.append(index)
```

### Extract the predictions given the indexes above

```python
with open("/Users/yuriyperezhohin2/Desktop/MastersAwareTech/Masters/ThesisImplementation/GitLab/sandabox-yuriy/VALUENET/outputTRAINVALUENET.txt", "r") as f:
    lines = f.readlines()
# Create a list of rows, where each row is a list of the three columns
rows = []
for line in lines:
    # Split the line on the tab character
    parts = line.split("\n")

    # Create a row from the three parts
    row = [parts[0]]

    # Add the row to the list of rows
    rows.append(row)

# Create a CSV writer object
with open("predicted_ValueNet.csv", "w", newline="") as f:
    writer = csv.writer(f)

    # Write the header row
    writer.writerow(["SQL"])

    # Write the data rows
    for row in rows:
        writer.writerow(row)
```

```python
predicted_dev = pd.read_csv("predicted_ValueNet.csv")
predicted_specific_databases= []
for index , row in predicted_dev.iterrows():
    if index in indexes:
        predicted_specific_databases.append(row["SQL"])
print(len(predicted_specific_databases))
predicted_specific_databases
```
